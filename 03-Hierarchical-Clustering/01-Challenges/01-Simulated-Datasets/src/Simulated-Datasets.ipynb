```python
# TODO: Imports
import numpy as np
from sklearn import datasets
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

### Q1. Load datasets from scikit-learn datasets using `make_circles`, `make_moons`

```python
# Load datasets noisy_circles, noisy_moons, and no_structure (ie random data)np.random.seed(42)
n_samples = 1500
noisy_circles = datasets.make_circles(n_samples=n_samples, factor=.5,noise=.05)
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)
no_structure = np.random.rand(n_samples, 2), None
```

```python
noisy_circles[0]
```

```python
# We give you this last dataset
np.random.seed(170)
X, y = datasets.make_blobs(n_samples=n_samples)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)
aniso = (X_aniso, y)
```

### Q3. Agglomerative Clustering on `noisy_circles`

```python
#TODO: Apply linkage to noisy_circles dataset and display dendrogram
linked = linkage(noisy_circles[0], method='ward') ## more balanced than single
dendrogram(linked)
plt.show()
```

```python
#TODO: Apply agglomerative clustering to the dataset
acc = AgglomerativeClustering(n_clusters=2, linkage='single')
```

```python
#TODO: Display clusters
labels = acc.fit_predict(noisy_circles[0])

plt.scatter(noisy_circles[0][:,0], noisy_circles[0][:,1], c=labels) #ok avec single // NOK avec ward
plt.show()
```

### Q4. Agglomerative Clustering on other datasets

```python
#TODO : repeat the same exercise on noisy_moons
acc = AgglomerativeClustering(n_clusters=2, linkage='single')

labels = acc.fit_predict(noisy_moons[0])

plt.scatter(noisy_moons[0][:,0], noisy_moons[0][:,1], c=labels) #ok avec single
plt.show()
```

```python
#TODO : repeat the same exercise on no_structure
acc = AgglomerativeClustering(n_clusters=8, linkage='ward')

labels = acc.fit_predict(no_structure[0])

plt.scatter(no_structure[0][:,0], no_structure[0][:,1], c=labels)
plt.show()
```
